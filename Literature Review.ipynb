**1.	Introduction**

The question of how people’s values relate to their political beliefs is important in the social sciences. In the current European context, this question has been tested by significant societal disruptions, most notably the COVID-19 pandemic and the rise of populist movements (Bonetto et al., 2021). Therefore, this project focuses on the Schwartz Theory of Basic Human Values to understand these dynamics, specifically analyzing how values such as universalism, security, and tradition predict political attitudes across different generations and contexts. The review aims to anchor the interpretations of the European Social Survey (ESS) data that were used in this project to track shifts in these relationships. The period covered by the review spans the pre-pandemic landscape, the subsequent recovery, and the new normal. Additionally, the review addresses key themes including the predictive power of specific values, generational differences, and the impact of collective shocks on value-driven political trust and policy preferences.

**2.	The Schwartz Theory of Basic Human Values**

Schwartz’s theory views values as important goals that shape how individuals see the world, make decisions, and act (Schwartz, 2021). The theory organizes ten basic values in a circular pattern, showing how they relate to one another. Values that are next to each other support each other, while those across from each other represent different priorities that can clash (Schwartz, 2021). The ten values, which are self‑direction, stimulation, hedonism, achievement, power, security, conformity, tradition, benevolence, and universalism, are measured in the ESS with the Portrait Values Questionnaire and have been shown to form a coherent structure across many European societies (Schwartz, Breyer, & Danner, 2015). The values are further grouped into four dimensions that capture broader motivational tensions central to social and political life. The values are grouped as follows: Self‑Transcendence (universalism and benevolence), Conservation (security, conformity, tradition), Self‑Enhancement (power, achievement), and Openness to Change (self‑direction, stimulation) (Schwartz, 2021). Moreover, the circular structure goes beyond just categorization; it represents motivational trade-offs that significantly influence social and political decisions (Schwartz, 2021). The Schwartz framework, therefore, provides a theoretically grounded and empirically validated lens for linking stable motivational orientations to both general ideological identification and concrete policy stances in European politics (Piurko, Schwartz, & Davidov, 2011).

**3.	Values as predictors of political attitudes in Europe**

Large multi‑country studies show that individuals who prioritise Self‑Transcendence, especially Universalism, are more favourable toward immigration, whereas those high on Conservation values express greater opposition to immigration across numerous European contexts, which establishes values as consistent psychological predictors of immigration attitudes (Davidov et al., 2020). A key factor is the perceived symbolic threat, which helps explain how Universalism or Conservation values influence attitudes toward immigration (Davidov et al., 2020). This shows that people’s values affect how they view cultural change and group boundaries. Evidence from the European Social Survey similarly indicates that Universalism is associated with lower perceived threat from immigration and less opposition to immigrant inflows, reinforcing the view that value priorities shape opinions on migration beyond sociodemographic factors (European Social Survey, 2021). According to Heath et al. (2020), an analysis of European public opinion shows that migration is a key issue that divides people’s beliefs. The authors further indicated that research that considers personal values accurately explains differences in opposition to migration across countries more than research that focuses only on economic interests (Heath et al., 2020). This further highlights the importance of using values to understand public attitudes in Europe. Davidov et al. (2020) concur that values do not simply correlate with immigration stances; they structure the threat narratives people apply to immigration, which is why they remain a robust starting point for modelling political attitudes in comparative research.

**3.1.	 Values and redistribution preferences**

Recent studies show that people who value Universalism and egalitarian principles are more likely to support income redistribution and solidarity within the EU. Furthermore, those who prioritize Security may also influence support for a European social benefit system, with variations across countries (Reinl, Seddig, Dennison, & Davidov, 2024). This evidence implies that moral concern that extends beyond the in‑group is a stable driver of solidaristic preferences, whereas preferences for order and stability, captured by Conservation and Security, can temper support for transnational redistribution (Reinl et al., 2024). While aggregate support for redistribution has been relatively stable over the years, there is substantial variation across welfare regimes and countries, suggesting that the context in which values are expressed shapes the extent to which they translate into policy preferences (Kudrnáč & Petrůšek, 2022). Enke, Rodríguez-Padilla, & Zimmermann (2023)’s focus on moral universalism shows that a universalist moral outlook predicts greater support for redistribution and other inclusive policies across Western democracies. This highlights a general mechanism that complements the Schwartz value approach. Concurrently, a focus on political economy by Cavaillé & Van der Straeten (2023) cautions that increasing diversity and the importance of immigration can influence people’s willingness to support solidarity. This helps explain why attitudes toward wealth redistribution vary across social classes and national contexts in Europe. These perspectives from the scholars indicate that Universalism tends to pull attitudes toward greater solidarity, whereas Security and related Conservation values can temper or redirect that impulse, with the strength of effects amplified in settings of higher material security and weakened where perceived risk or out‑group salience is high. 

**3.2.	 From values to left-right self‑placement**

The relationship between values and political beliefs corresponds with the Green Alternative Libertarian (GAL) and Traditional Authoritarian Nationalist (TAN) framework by Crulli & Emanuele (2025). Values such as Universalism and Benevolence lead people to support GAL views, while values such as Security, Conformity, and Tradition push them toward TAN views. This helps explain where people fall on the left-right political spectrum in modern Europe. Lönnqvist & Ilmarinen (2025) posit that Universalism is linked to the cultural GAL-TAN axis and somewhat to the economic left-right spectrum in voting and ideology. In contrast, Security is associated with support for populist options, showing that core values influence policy views and voting behaviour (Lönnqvist & Ilmarinen, 2025). Most European studies use a left-right self-placement scale to measure ideology (Aybar, Pérez, & Pavía, 2024). Therefore, it is important to note recent evidence that scale design can affect placement distributions, which motivates careful testing of which specific values have the largest predictive power for the left-right self‑placement scale from 2018 to 2024, along with sensitivity checks on measurement choices. The literature indicates that Universalism and Benevolence typically push respondents leftward, whereas Conformity and Tradition pull them rightward, and that these relationships depend on context.

**4.	Generational structuring of values and attitudes pre‑ and post‑pandemic**

Generational analysis rests on the idea that people’s core values are shaped in their formative years, generally late adolescence to the mid‑twenties, after which they tend to stabilise and change slowly (Welzel et al., 2024). Research conducted during and after the pandemic supports this logic by showing that values are usually stable but can undergo short‑term adjustments when society experiences major disruption (Welzel et al., 2024). This means cohorts exposed to different historical contexts often develop distinct value profiles that can persist as they age (O’Grady, 2023). The COVID‑19 shock provides a clear illustration of how a shared event can nudge value priorities, with studies during the first lockdowns reporting higher Conservation and lower Openness to change in the short run (Bonetto et al., 2021). The study by Vecchione (2022) also shows meaningful individual‑level differences in value change, with worsening personal finances linked to rising Conservation, indicating that shocks interact with life stage and circumstances rather than reshaping values uniformly across all young people. Research on European youth shows that cohort differences in political positioning exist, but the size and direction of these gaps vary across countries (Grassi, Portos, & Felicetti, 2024). Other studies likewise finds limited evidence of growing age‑based polarisation across Europe and points instead to persistent cohort effects that depend on period conditions, not a single pan‑European shift among the young (O’Grady, 2023). The literature indicates that formative experiences attach different political meanings to the same institutions across countries, creating cohort gaps that reflect context‑by‑cohort interactions (Lauterbach & De Vries, 2021). Moreover, patterns of youth engagement differ with the institutional environment, Kitanova (2020) argue that cross‑national analyses show higher conventional participation among young people in more established democracies, suggesting similar values can translate into different behaviours depending on opportunity structures. Grassi, Portos, & Felicetti (2024) purport that both shocks and institutional settings shape how generational value profiles map onto politics. These studies not only emphasise the need for a generational lens that is sensitive to national context but also they motivate analysing values and attitudes by cohort before and after the pandemic while allowing for country heterogeneity.

**5.	Pandemic shock, Conservation, and political trust**

Evidence from France during the first lockdown indicates a rise in Conservation values and a drop in Openness to Change (Bonetto et al., 2021). This pattern supports the hypothesis that external shocks can temporarily shift motivational priorities toward security and order, even if long‑term value baselines remain relatively stable (Bonetto et al., 2021). A study from Italy found that while average higher-order values did not change from mid to late 2020, individual experiences varied widely, however income loss to increased Conservation, indicating that economic strain can influence values for certain groups (Vecchione, 2022). Evidence from Poland documents increases in self‑direction, security, conformity, humility, caring, and universalism, with hedonism declining (Bojanowska et al., 2021). A study by Iosifyan et al. (2023) on perceived threats to values further shows that people saw COVID‑19 as endangering personal‑focused values more than social‑focused ones, and that stronger Conservation was associated with protective behaviours and heightened fear. These views indicate short‑run flexibility, as Conservation can rise and Openness can soften during periods of acute uncertainty, though the direction and magnitude of change vary by country, time window, and individual exposure to risk. A comparative study across twelve countries, including France, Germany, and Italy, shows that trust in scientists was the key driver of support for and compliance with non‑pharmaceutical interventions, whereas trust in government had more ambiguous effects when political and scientific messages diverged (Algan et al., 2021). Similar to the survey study by Bicchieri et al. (2021), which collected evidence across nine countries, including Spain and the United Kingdom, finds that compliance depends on empirical and normative expectations, and that these expectations work more strongly when trust in science and government is higher. Stivas & Cole (2023) assert that in sixteen European Union (EU) member states, effective pandemic management is associated with higher political trust and greater transparency, underscoring the governance channels through which trust shapes outcomes. However, Seyd & Bu (2022) insist that the effect of trust on compliance in Austria, Germany, and the United Kingdom strengthened as fear of infection waned, indicating a time‑varying role of trust across pandemic stages. Across these perspectives, a consistent idea is that political and scientific trust align with higher compliance and better evaluations of government performance, whereas generalised social trust can be ambiguous if it reduces perceived need for caution because people expect others to self‑police. 

**6.	Universalism, redistribution, and generational change**

Value profiles anchored in Universalism and the broader Self‑Transcendence cluster are positively associated with support for redistribution and cross‑border solidarity instruments (Reinl, Seddig, Dennison, & Davidov, 2024). Universalism predicts endorsement of an EU wide social benefit scheme, which means that people who value solidarity are more likely to back extensive welfare programs that go beyond national borders (Reinl et al., 2024). A focus on political economy shows that believing in universalism morals strongly influences the desire for wealth redistribution in Western democracies (Enke, Rodríguez‑Padilla, & Zimmermann, 2023). This suggests a common underlying reason that aligns with the Schwartz framework and applies to different issues. Furthermore, value policy correlations for multiculturalism and redistribution mirror the Universalism pattern, which strengthens the case that Self‑Transcendence is a consistent psychological base for democratic preferences among European voters (Lönnqvist & Ilmarinen, 2025). The literature on welfare chauvinism shows that support for redistribution can become conditional when beneficiaries are perceived as outgroups, with publics generally more restrictive toward immigrants’ access to cash benefits than to in‑kind services (Eick, Larsen, Geiger, & Sundberg, 2021). According to Kros & Coenders (2019), increases in perceived ethnic threat over time raise welfare‑chauvinist attitudes even after accounting for egalitarian orientations. This indicates that cultural threat can override universalistic predispositions when immigration becomes salient. Furthermore, welfare chauvinism is now embedded in both party competition and mass attitudes across Europe, suggesting that exclusionary logics have become a durable constraint on translating universalist values into unconditional welfare support (Careja & Harris, 2022). These constraints are evident in universal basic income proposals, as surveys show that a simple immigration cue can significantly erode initial support among prior supporters (Bay, Hellevik, & Pedersen, 2024). Similarly, recent research finds that short‑term increases in regional immigration can shift attitudes negatively, even when stock levels are high, suggesting a temporal salience channel that can destabilise solidaristic support during inflow spikes (Dražanová & Gonnot, 2023). Therefore, the link between universalism and redistribution is strong in theory, but its real-world application depends on program design and immigration issues that trigger welfare chauvinism. This is why it is important to re-test Gen Z, who tend to have more cosmopolitan views, to see if their support for solidarity changes when immigration becomes a political issue.

**7.	Cross‑national heterogeneity in value-attitude correlations**

Across several European studies, such as that of Davidov et al. (2020) and Kudrnáč & Petrůšek (2022), the sign of value effects is remarkably consistent, Self‑Transcendence is associated with lower threat and greater support for immigration, while Conservation aligns with greater opposition, but the magnitude of these relationships varies substantially across countries. When moving from immigration attitudes to redistribution, the pattern also holds on average, Universalism predicts stronger support for redistributive and solidarity policies, whereas Conservation is linked to restraint (Reinl et al., 2024; Kudrnáč & Petrůšek, 2022). Likewise, Lönnqvist & Ilmarinen (2025) found that when mapping values to parties, ideology, and policy, there are stronger value politics associations in Western Europe than in other regions, illustrating regional unevenness in how values translate into political stances. In a cross‑national analysis of support for an EU wide social benefit, the country’s net contribution to the EU budget strengthens or weakens the impact of Universalism on solidarity preferences, which indicates that material and institutional environments filter value expression (Reinl et al., 2024). 

**8.	Research Gap**

Studies such as those of Davidov et al. (2020) and Reinl, Seddig, Dennison, & Davidov (2024) have conducted extensive work linking Schwartz values to immigration, redistribution, and ideology in Europe using 2014/2015 data from the immigration module of the European Social Survey. However, there is a lack of post‑pandemic, cross‑national assessment of how specific basic values such as Universalism and Security map onto left-right self‑placement for Gen Z and Millennials compared to older cohorts using the same measurement framework across ESS Rounds 9, 10, and 11. Bonetto et al. (2021) and Vecchione (2022) showed that COVID‑19 temporarily tilted value priorities toward Conservation and away from Openness, but they do not establish whether these shifts produced durable cohort differences in ideological positioning across countries. These gaps in the literature concur the need for a project that will investigate the evolving relationship between personal values and political identity across generations in Europe from 2018 to 2024.

**9.	Conclusion**

The relationship between personal values and political identity in Europe is dynamic and multifaceted. Between 2018 and 2024, Schwartz’s basic human values, particularly Universalism and Security, have remained fundamental anchors for political self-placement and voting behaviour. While younger generations like Gen Z exhibit a more progressive value profile often driven by higher education levels, these trends are moderated by national context and the enduring influence of Conservation values among older cohorts. The COVID-19 pandemic served as a significant shock that influenced political trust and specific policy preferences, but it did not lead to a wholesale, measurable shift in the core human values of the European population. Methodological rigor, particularly regarding measurement invariance, remains essential for future research to accurately track these evolving relationships in a complex and changing Europe.
